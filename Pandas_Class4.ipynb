## 1. Boolean Indexing & Conditional Selection
- Use comparison operators to build conditions (e.g., `df['col'] > 50`).
- Pass the condition inside `df[...]` to get matching rows.
- Combine multiple conditions with `&` (AND), `|` (OR), and `~` (NOT).
- Parentheses are required when combining conditions.

```python
import pandas as pd
import numpy as np 
# Same dataset
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)
print("Original DataFrame:")
print(b)

# 1. Single condition: rows where Age > 30
print("\nRows where Age > 30:")
print(b[b["Age"] > 30])

# 2. Single condition: rows where Score < 90
print("\nRows where Score < 90:")
print(b[b["Score"] < 90])

# 3. Multiple conditions (AND): Age > 25 AND Score > 85
print("\nRows where Age > 25 AND Score > 85:")
print(b[(b["Age"] > 25) & (b["Score"] > 85)])

# 4. Multiple conditions (OR): Age < 25 OR Score > 90
print("\nRows where Age < 25 OR Score > 90:")
print(b[(b["Age"] < 25) | (b["Score"] > 90)])

# 5. Using NOT (~): rows where Score is NOT greater than 85
print("\nRows where Score is NOT greater than 85:")
print(b[~(b["Score"] > 85)])
```

## 2. Filtering Rows with Multiple Conditions
- Combine two or more conditions to refine selection.
- Remember to enclose each condition in parentheses.

```python
# 1. Age > 25 AND Score > 85
print("\nRows where Age > 25 AND Score > 85:")
print(b[(b["Age"] > 25) & (b["Score"] > 85)])

# 2. Age < 35 OR Score >= 90
print("\nRows where Age < 35 OR Score >= 90:")
print(b[(b["Age"] < 35) | (b["Score"] >= 90)])

# 3. Age between 25 and 35 (inclusive)
print("\nRows where Age is between 25 and 35 (inclusive):")
print(b[(b["Age"] >= 25) & (b["Age"] <= 35)])
```

## 3. Using `query()` for Cleaner Filtering
- `query()` allows filtering rows with a string expression.
- It's concise and readable for complex conditions.
- Variables can be passed using the `@` symbol.

```python
# 1. Simple query: Age > 25
print("\nRows where Age > 25:")
print(b.query("Age > 25"))

# 2. Multiple conditions: Age < 35 and Score >= 85
print("\nRows where Age < 35 and Score >= 85:")
print(b.query("Age < 35 and Score >= 85"))

# 3. Using a Python variable inside query (with @)
threshold = 90
print(f"\nRows where Score > {threshold}:")
print(b.query("Score > @threshold"))
```

## 4. String Operations on Columns
- Use the `.str` accessor for vectorized string methods.
- Common methods: `contains`, `replace`, `startswith`, `endswith`, `lower`, `upper`.
- Useful for cleaning or filtering text data.

```python
# 1. Convert names to lowercase
print("\nNames in lowercase:")
print(b["Name"].str.lower())

# 2. Convert names to uppercase
print("\nNames in uppercase:")
print(b["Name"].str.upper())

# 3. Check if names contain the letter 'a' (case-insensitive)
print("\nRows where Name contains 'a':")
print(b[b["Name"].str.contains("a", case=False)])

# 4. Replace a substring (change 'Charlie' to 'Charles')
c = b.copy()
c["Name"] = c["Name"].str.replace("Charlie", "Charles", regex=False)
print("\nAfter replacing 'Charlie' with 'Charles':")
print(c)

# 5. Filter rows where Name starts with 'D'
print("\nRows where Name starts with 'D':")
print(b[b["Name"].str.startswith("D")])

# 6. Filter rows where Name ends with 'e'
print("\nRows where Name ends with 'e':")
print(b[b["Name"].str.endswith("e")])
```

## Mini Practice
1. Build a DataFrame with columns such as Name, Age, City, Score.
2. Use boolean indexing to select rows where Age > 25.
3. Filter rows where City is either 'Delhi' or 'Mumbai'.
4. Use query() to find rows with Score >= 85 and Age < 35.
5. Clean a text column using str.replace or str.lower.

```python
import pandas as pd

# 1. Build a DataFrame with columns: Name, Age, City, Score
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma", "Frank"],
    "Age": [24, 32, 28, 40, 22, 30],
    "City": ["Delhi", "Mumbai", "Pune", "Chennai", "Delhi", "Mumbai"],
    "Score": [90, 82, 88, 75, 95, 84]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# 2. Use boolean indexing to select rows where Age > 25
age_filter = df[df["Age"] > 25]
print("\nRows where Age > 25:")
print(age_filter)

# 3. Filter rows where City is either 'Delhi' or 'Mumbai'
city_filter = df[df["City"].isin(["Delhi", "Mumbai"])]
print("\nRows where City is Delhi or Mumbai:")
print(city_filter)

# 4. Use query() to find rows with Score >= 85 and Age < 35
query_filter = df.query("Score >= 85 and Age < 35")
print("\nRows with Score >= 85 and Age < 35:")
print(query_filter)

# 5. Clean a text column using str.replace or str.lower
# Example: convert City to lowercase and replace 'delhi' with 'New Delhi'
cleaned = df.copy()
cleaned["City"] = cleaned["City"].str.lower()           # to lowercase
cleaned["City"] = cleaned["City"].str.replace("delhi", "New Delhi", regex=False)
print("\nAfter cleaning City column:")
print(cleaned)
```
