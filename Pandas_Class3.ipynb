## Handling Missing Values
Real-world datasets often have missing data. Pandas offers:
- `isna()` / `isnull()` â†’ detect missing values (returns True/False)
- `dropna()` â†’ remove missing values (rows/columns)
- `fillna()` â†’ replace missing values with a given value or method (e.g., forward fill)

Always inspect how much data is missing before deciding to drop or fill.

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with some missing values
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, None, 35, 40, None],
    "City": ["Delhi", "Mumbai", None, "Bangalore", "Chennai"],
    "Score": [85, 91, np.nan, 88, 95]
}

b = pd.DataFrame(a)
print("Original DataFrame:")
print(b)

# 1. Detect missing values
print("\nCheck where data is missing:")
print(b.isna())

print("\nCount of missing values in each column:")
print(b.isna().sum())

# 2. Drop rows with any missing value
c = b.dropna()
print("\nAfter dropping rows with missing values:")
print(c)

# 3. Fill missing values
# Fill numeric columns with mean, text columns with a placeholder
d = b.copy()
d["Age"] = d["Age"].fillna(d["Age"].mean())
d["City"] = d["City"].fillna("Unknown")
d["Score"] = d["Score"].fillna(0)
print("\nAfter filling missing values:")
print(d)
```

## Changing Data Types (`astype`)
- Convert a column to a different type using `astype()`.
- Examples: converting strings to numbers, integers to floats, or columns to category.

Be careful: make sure the data is compatible with the new type!

```python
import pandas as pd

# Same dataset
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)
print("Original DataFrame:")
print(b)

# 1. Convert "Age" column from int to float
b["Age"] = b["Age"].astype(float)
print("\nAge column converted to float:")
print(b.dtypes)

# 2. Convert "Name" column to category type (memory-efficient)
b["Name"] = b["Name"].astype("category")
print("\nName column converted to category:")
print(b.dtypes)

# 3. Convert "Score" column to string
b["Score"] = b["Score"].astype(str)
print("\nScore column converted to string:")
print(b.dtypes)
```

## Replacing Values (`replace`)
- Use `replace()` to substitute specific values or patterns.
- Useful for cleaning labels, fixing typos, or mapping categories.

```python

# 1. Replace a single value in a column
c = b.copy()
c["Name"] = c["Name"].replace("Bob", "Robert")
print("\nAfter replacing 'Bob' with 'Robert':")
print(c)

# 2. Replace multiple values at once
d = b.copy()
d["Score"] = d["Score"].replace({88: 90, 95: 99})
print("\nAfter replacing 88 -> 90 and 95 -> 99 in Score:")
print(d)

# 3. Replace using a pattern in strings (e.g., shorten names)
e = b.copy()
e["Name"] = e["Name"].replace({"Charlie": "Ch"}, regex=False)
print("\nAfter shortening 'Charlie' to 'Ch':")
print(e)
```

## Dropping Rows & Columns (`drop`)
- `drop(labels, axis=0)` â†’ drop rows by index labels.
- `drop(labels, axis=1)` â†’ drop columns by name.
- Use `inplace=True` if you want to modify the DataFrame directly.

```python

# 1. Drop a single row by index label
c = b.drop(2)   # drops row with index 2 (Charlie)
print("\nAfter dropping row with index 2:")
print(c)

# 2. Drop multiple rows
d = b.drop([0, 4])   # drops Alice and Emma
print("\nAfter dropping rows 0 and 4:")
print(d)

# 3. Drop a single column
e = b.drop("Score", axis=1)
print("\nAfter dropping the 'Score' column:")
print(e)

# 4. Drop multiple columns
f = b.drop(["Age", "Score"], axis=1)
print("\nAfter dropping 'Age' and 'Score' columns:")
print(f)

# 5. Using inplace=True to modify the DataFrame directly
g = b.copy()
g.drop("Score", axis=1, inplace=True)
print("\nAfter dropping 'Score' with inplace=True:")
print(g)
```

## Detecting & Removing Duplicates
- `duplicated()` â†’ returns a Boolean Series marking duplicate rows.
- `drop_duplicates()` â†’ remove duplicate rows.
- You can specify `subset` (columns) and `keep` ('first', 'last', or False).

```python
import pandas as pd

# Create a custom dataset with some duplicate rows
data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "Bob", "David"],
    "Age": [25, 30, 35, 25, 30, 40],
    "Score": [88, 92, 79, 88, 92, 85]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# 1. Detect duplicates
print("\nBoolean Series showing duplicates:")
print(df.duplicated())

# 2. Count total duplicates
print("\nTotal number of duplicate rows:")
print(df.duplicated().sum())

# 3. Remove duplicates (keep first occurrence)
df_first = df.drop_duplicates()
print("\nAfter dropping duplicates (keep='first'):")
print(df_first)

# 4. Remove duplicates (keep last occurrence)
df_last = df.drop_duplicates(keep="last")
print("\nAfter dropping duplicates (keep='last'):")
print(df_last)

# 5. Remove all duplicates (keep=False)
df_none = df.drop_duplicates(keep=False)
print("\nAfter removing all duplicates (keep=False):")
print(df_none)

# 6. Remove duplicates based on a subset of columns (e.g., Name only)
df_subset = df.drop_duplicates(subset=["Name"])
print("\nAfter dropping duplicates based on 'Name' only:")
print(df_subset)
```
